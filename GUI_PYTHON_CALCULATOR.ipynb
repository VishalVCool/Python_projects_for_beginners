```python
from tkinter import *
def click(event):
    global screenvalue
    text = event.widget.cget("text")
    print(text)
    if text=="=":
        if screenvalue.get().isdigit():
            value=int(screenvalue.get())             # here we get value from screenvalue variable
        else:
            value=eval(screen.get())                 #eval evaluate strings example 2*5 =10
            screenvalue.set(value)
            screen.update()
            
    elif text == "c":                               #here for button c we say clear whole screen in one click
        screenvalue.set("")                         # setting screen to empty
        screen.update()
    else:
        screenvalue.set(screenvalue.get() + text)   #incase button is not c proceed further
        screen.update()
        
        
root=Tk()

root.geometry("380x600")
root.minsize(380,600)
root.maxsize(380,600)


root.title("Calculator")
l1=Label(text="This GUI calculator is written in python", font="tribubol 8 ")
l1.pack(side=BOTTOM)

screenvalue=StringVar()     #screenvalue variable will hold stringvalue
screenvalue.set("")         # we set screen value to empty by default

frame1=Frame(root, borderwidth=0.5, bg="black", relief=SUNKEN)
screen = Entry(frame1, textvariable=screenvalue, font="tribubol 50 bold")           #we created an Entry widget here
screen.pack(fill=X, padx=2, pady=2)
frame1.pack(fill=X, padx=2, pady=2)

frame2=Frame(root, borderwidth=1)           #frame of buttons 789

b1=Button(frame2, text="7", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2,)
b1.bind("<Button-1>",  click)

b1=Button(frame2, text="8", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame2, text="9", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame2, text="/", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

frame2.pack(fill=X)

frame3=Frame(root, borderwidth=1)                #frame of buttons 456

b1=Button(frame3, text="4", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2,)
b1.bind("<Button-1>",  click)

b1=Button(frame3, text="5", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame3, text="6", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame3, text="*", font="tribubol 30",padx=17, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

frame3.pack(fill=X)

frame4=Frame(root, borderwidth=1)                #frame of buttons 123

b1=Button(frame4, text="1", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2,)
b1.bind("<Button-1>",  click)

b1=Button(frame4, text="2", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame4, text="3", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame4, text="-", font="tribubol 30",padx=18, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

frame4.pack(fill=X)

frame5=Frame(root, borderwidth=1)                #frame of buttons 0.%

b1=Button(frame5, text="0", font="tribubol 30",padx=20, pady=10)
b1.pack(side=LEFT, padx=2,)
b1.bind("<Button-1>",  click)

b1=Button(frame5, text=".", font="tribubol 30",padx=25, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame5, text="%", font="tribubol 30",padx=14, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

b1=Button(frame5, text="+", font="tribubol 30",padx=13, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)

frame5.pack(fill=X)

frame6=Frame(root, borderwidth=1)                #frame of buttons c =

b1=Button(frame6, text="c", font="tribubol 30",padx=70, pady=10)
b1.pack(side=LEFT, padx=2,)
b1.bind("<Button-1>",  click)

b1=Button(frame6, text="=", font="tribubol 30",padx=70, pady=10)
b1.pack(side=LEFT, padx=2)
b1.bind("<Button-1>", click)
frame6.pack(fill=X)

root.mainloop()
```
